# Outlier handeling

```python
# Calculate the 25th (Q1) and 75th (Q3) percentiles for the 'MaxTemp' column.

Q1 = df['MaxTemp'].quantile(0.25)
Q3 = df['MaxTemp'].quantile(0.75)
```

```python
print(Q1,Q3)
```

```python
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df['MaxTemp'] = np.where(df['MaxTemp'] > upper_bound, upper_bound, df['MaxTemp'])
df['MaxTemp'] = np.where(df['MaxTemp'] < lower_bound, lower_bound, df['MaxTemp'])
```

# My approch of solving the task

1) Importing data: Importing csv file given to the jupyter notebook is the foremost thing to do because with this data only further data frame is created to do further calculations of ml.

2) Understanding data: First we need to understand the data by analyzing data because this helps in getting better output by avoiding possible errors.
This we can do with the help of df.info() which provides informaition of all columns of dataframe like number of rows, data type etc which helps in understanding data.

3) Data preprocessing: This is the crucial part of the process. This is the step which needs to be done before applying models. And there are few steps    to follow they are:
   * Outlier handeling: Outliers exist because of possible human error while uploding data. We can find these outlier boxplot, Scatter plot etc(I have 
     used boxplot for this project) to visually find outliers. After knowing variables or columns which has outlier we handle them either using capping 
     method or IQR-based filtering. I have used IQR filtering in this project to handel outliers which manages to keep the data within the normal range.

   * Creating dummy variables : we create dummy values to the variables which are categorical type but are important using get_dummies function. This        will create column for each category of given column and put the values accordingly(like 0 and 1). In this project rain today and tomorrow columns      has boolean type that is yes/no so we create 0s and 1s for each new columns and we delete extra columns which are not usefull. This is important        because we cannot calculate or apply ML models to categorical or non numeric columns.

   * Missing value handling: There exists some missing value or null value in dataset so we handle this issue by adding appropriate values like              mean,mode values in missing places to get proper output and maintain proper range of data.

   * Variable transformation : We find variables which are not so usefull in predicting rainfall then we delete it so that we can achive less                computational cost

4) X y split: We need to create X variable and y because X refers to the input features or independent variables (e.g.cloud, temperature) used to        make predictions.y refers to the target variable or dependent variable (In this project rain tomorrow), which the model aims to predict based on X.

5) Train test split: This is the step where we split our dataset into train dataset and test data set initially we take test size as 0.2 means 20% test dataset and remaining 80% is for training data set this is the ideal situation. This is important because we train the train dataset to test test dataset and get accuracy score.

6) Applying classification models to train data set: We do this to train data for training set and the classification models used are:
   * Logistic regression
   * Linear Discriminant Analysis (LDA)
   * K-Nearest Neighbors (KNN)
   * Classification tree

    
    

    For better performance we also use ensemble techniques they are:
   
       * Bagging
       * Random forest 
       * Gradient boost
       * ADA boost
       * XG boost

7) Predict y variable: With the help of classification models we predict both train data and test data. But here predicting test data is important because model is predicting for unknown data which will help us to know whether it is correctly predicting or not.

8) Check accuracy: Through confusion matrix or accuracy score we can know how accuratly our model is predicting.

9) Compare accuracy scores of test data with train data: If difference beteween accuracy scores of predicted train and test is more it means there exist  overfitting.

10) Handle overfitting: There are many methods to handle including ensemble methods they are:
    * Regularization: Apply techniques like L1 or L2 regularization to the classifier (e.g., Logistic Regression, SVM) to constrain the model's               complexity.
    * Prune the Model: For decision trees, limit the maximum depth, minimum samples per leaf, or minimum samples per split.
    * Cross-Validation: Use cross-validation to evaluate the model's performance on different subsets of data to ensure it generalizes well.
    * Feature Selection: Remove irrelevant or redundant features to reduce the model's complexity.
    * Increase Data: Gather more data or use data augmentation techniques to improve generalization.
    * Ensemble Methods: Use techniques like bagging or boosting to improve stability and reduce overfitting.

I have used only pruning models, ensemble methods and hyperparameter tuninig

11) Choose best model: choose the model whose accuracy score is high and it is consistent with different data sets.
